```python
import hashlib
  
# initializing of string
str2hash = "Hello World!"
  
# encoding Hello World! using encode() then sending to md5()
result = hashlib.md5(str2hash.encode())
  
# print Equivalent hexadecimal value.
print("The hexadecimal equivalent of hash is : ", end ="")
print(result.hexdigest())
 
# print Equivalent byte value.
print("The byte equivalent of hash is : ", end ="")
print(result.digest())
 
 
#print clone of the hash object.
print("The clone of hash object is : ",end="")
print(result.copy())
```
